1. What is the difference between a function and a method in Python?

Ans:- Function: A standalone block of code that performs a specific task. It's defined with def and called directly.

Method: A function defined within a class, associated with an object. It operates on that object's data and is called using the object (e.g., object.method()).

Key Difference:

The key difference is that a method is associated with an object and operates on that object's data, while a function is independent and can be called directly. You can see in the example that the method bark() is called using the object my_dog, while the function greet() is called directly.

Example:- # Function
def greet(name):
  print(f"Hello, {name}!")

# Method (within a class)
class Dog:
  def bark(self):
    print("Woof!")

# Using the function
greet("Alice")  # Output: Hello, Alice!

# Using the method
my_dog = Dog()
my_dog.bark()  # Output: Woof!

2. Explain the concept of function arguments and parameters in Python?

Ans:-Parameters:

Parameters are the variables listed inside the parentheses in the function definition.
They act as placeholders for the values that will be passed to the function when it's called.
Arguments:

Arguments are the actual values passed to a function when it's called.
They are assigned to the corresponding parameters in the function definition.
In simpler terms: Parameters are like labeled boxes within a function, and arguments are the items you put into those boxes when you use the function.

Example:-

def greet(name, age):  # 'name' and 'age' are parameters
  print(f"Hello, {name}! You are {age} years old.")

greet("Alice", 30)  # "Alice" and 30 are arguments

In this example:

name and age are parameters defined in the greet function.
When we call greet("Alice", 30), "Alice" and 30 are arguments passed to the function.
Inside the function, "Alice" is assigned to the name parameter, and 30 is assigned to the age parameter.
Types of Arguments:

Positional Arguments: Arguments passed in the order they are defined in the function.
Keyword Arguments: Arguments passed with their parameter names (e.g., age=30). This allows for flexibility in the order of arguments.

Default Arguments: Values assigned to parameters in the function definition. If an argument isn't provided for that parameter, the default value is used.

4. What is the purpose of the `return` statement in a Python function?

Ans:-The return statement is used to:

Exit the function: When the return statement is encountered, the function immediately stops executing and returns control to the caller.

Send back a value (optional): The return statement can also be used to send a value back to the part of the program that called the function. This value can be of any data type (e.g., integer, string, list, etc.).

Example:-

  """Adds two numbers and returns the result."""

5. What are iterators in Python and how do they differ from iterables?

Ans:-Iterables:

Definition: An iterable is any Python object capable of returning its members one at a time, allowing it to be used in a for loop.

Examples: Lists, tuples, strings, dictionaries, files.

Mechanism: Iterables have an __iter__ method that returns an iterator object. This iterator is used to fetch the next item in the sequence.
Iterators:

Definition: An iterator is an object that represents a stream of data. It returns the data one element at a time using the __next__ method.

Purpose: To loop over iterables efficiently without loading the entire data into memory at once. They are used to retrieve the next item in the sequence.

Functionality:

Calling next() or __next__ on an iterator returns the next item in the sequence.

When there are no more items, it raises a StopIteration exception.

The main distinctions:

Purpose: Iterables are containers for data; iterators retrieve data from iterables.

Method: Iterables have __iter__; iterators have __next__.

Usage: You loop directly over iterables in for loops; you obtain an iterator from an iterable using iter() and then access elements with next().

Memory: Iterables might load all data into memory initially, while iterators load elements as needed, saving memory.

Example:-

my_list = [1, 2, 3] # This is an iterable

# Get an iterator from the list:
my_iterator = iter(my_list)

# Access elements using next():
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
#print(next(my_iterator))  # Raises StopIteration (no more items)
my_list = [1, 2, 3] # This is an iterable

6. Explain the concept of generators in Python and how they are defined?

Ans:-Okay, let's delve into the concept of generators in Python:

What are Generators?

Generators are a special type of function in Python that produce a sequence of values one at a time, instead of computing them all at once and storing them in memory. They are essentially iterators created using functions.

How They Work:

Generators use the yield keyword instead of return to produce a value.
When a generator function is called, it doesn't execute the function body immediately. Instead, it returns a generator object.
Each time you call next() on the generator object, it executes the function body until it encounters a yield statement.
The value specified by yield is returned, and the generator's state is saved.
On subsequent calls to next(), the generator resumes execution from where it left off, continuing until it encounters another yield or the end of the function.
Defining Generators:

You define a generator function just like a regular function, but instead of using return, you use yield to return values one by one.

Example:-

def my_generator():
  for i in range(5):
    yield i  # Yields each value in the range

# Create a generator object
gen = my_generator()

# Access values using next()
print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
# ... and so on

Benefits of Generators:

Memory Efficiency: Generators produce values on demand, saving memory compared to storing an entire sequence in memory.
Lazy Evaluation: Values are generated only when needed, improving performance.
Readability: Generators can make code more concise and readable for tasks involving sequences

In essence:

Generators are a powerful tool for creating iterators in a concise and memory-efficient way. They are particularly useful when dealing with large datasets or infinite sequences.

7. What are the advantages of using generators over regular functions?

Ans:- 1. Memory Efficiency

Generators: Produce values one at a time, on demand. They don't store the entire sequence in memory, making them highly memory-efficient, especially for large datasets or infinite sequences.

Regular Functions: Typically compute and return the entire result at once, which can consume significant memory if the result is large.

2. Lazy Evaluation:

Generators: Values are generated only when needed (lazy evaluation). This can improve performance, especially when you don't need to process all the values in a sequence.

Regular Functions: Execute the entire function body and return the result, even if you only need a portion of it.

3. Readability and Conciseness:

Generators: Can often express iterative processes more concisely and readably than regular functions, particularly when dealing with complex sequences.

Regular Functions: May require more code to achieve the same result as a generator, especially for tasks involving iteration.

4. Representing Infinite Sequences:

Generators: Can represent infinite sequences, as they generate values on demand without storing them all.

Regular Functions: Cannot directly represent infinite sequences, as they would require infinite memory to store the entire sequence.

5. Pipelining Generators:

Generators: Can be chained together (pipelined) to create complex data processing workflows, where the output of one generator becomes the input of another.

Regular Functions: Typically require intermediate storage of results, which can add complexity and reduce efficiency.

In summary:

Generators offer significant advantages over regular functions in terms of memory efficiency, lazy evaluation, readability, and the ability to represent infinite sequences. They are a powerful tool for creating iterators and processing large datasets in Python.

A lambda function is a small, anonymous function (a function without a name) defined using the lambda keyword. They are essentially throwaway functions that you can define and use inline within your code.

Flexibility: They can be easily passed as arguments to other functions, promoting code reusability and functional programming paradigms.

Limitations:

Limited to a single expression: Lambda functions cannot contain multiple statements or complex logic.

Best for simple tasks: They are ideal for short, throwaway functions, not for complex operations that require multiple steps or extensive logic.

9. Explain the purpose and usage of the `map()` function in Python?

Ans:-
Purpose:

The map() function applies a given function to each item of an iterable (e.g., a list, tuple, etc.) and returns an iterator that yields the results. It's a way to process and transform all elements of an iterable without using explicit loops.

Usage:

The basic syntax of map() is:
map(function, iterable)

function: The function to be applied to each element of the iterable.

iterable: The iterable containing the elements to be processed.

How it Works:

The map() function takes a function and an iterable as arguments.

It iterates over each element of the iterable.

For each element, it calls the provided function with the element as an argument.

The results of the function calls are yielded one by one, creating a new iterator.

Example:-

numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

1.lambda x: x**2 is a lambda function that squares a number.

2.map() applies this function to each element in the numbers list.

3.The results are collected into a new list called squared_numbers

Benefits of Using map():

Concise Code: It allows you to express transformations on iterables in a compact way, avoiding explicit loops.

Readability: It can improve code readability by clearly separating the transformation logic from the iteration.

Efficiency: map() can be more efficient than explicit loops in some cases due to its underlying implementation.

Important Considerations:

map() returns an iterator in Python 3. To get the results as a list, you need to use list(map(...)).

The function passed to map() should take a single argument (corresponding to an element from the iterable).

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Purpose: Applies a function cumulatively to the items of an iterable, reducing it to a single value.

Input: A function and an iterable.

Output: A single accumulated value.

    numbers = [1, 2, 3, 4, 5]

Input: A function and an iterable.

Output: An iterator containing the filtered elements.

Example:-

numbers = [1, 2, 3, 4, 5]

```
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
#the list?

#Answer:-

def sum_of_even_numbers(numbers):
  """
  Calculates the sum of all even numbers in a list.

  Args:
    numbers: A list of numbers.

  Returns:
    The sum of all even numbers in the list.
  """
  sum_even = 0
  for number in numbers:
    if number % 2 == 0:  # Check if the number is even
      sum_even += number  # Add even number to the sum
  return sum_even

# Example usage
numbers = [1, 2, 3, 4, 5, 6]
even_sum = sum_of_even_numbers(numbers)
print(f"The sum of even numbers in the list is: {even_sum}")
#To see the output, run the code.
```

```
#4. Write a Python function that checks if a given number is prime or not from 1 to 200./
#Answer
def is_prime(number):
  """
  Checks if a given number is prime.

  Args:
    number: The number to check.

  Returns:
    True if the number is prime, False otherwise.
  """
  if number <= 1:
    return False  # Numbers less than or equal to 1 are not prime
  for i in range(2, int(number**0.5) + 1):
    if number % i == 0:
      return False  # If divisible by any number, it's not prime
  return True  # If not divisible by any number, it's prime

# Example usage to check numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num)
```

```
##5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#terms?
##Answer
class FibonacciIterator:
    """
    An iterator class for generating the Fibonacci sequence.
    """

    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.current_term = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term < self.max_terms:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.current_term += 1
            return result
        else:
            raise StopIteration

# Example usage
fib_iterator = FibonacciIterator(10)  # Generate up to 10 terms
for number in fib_iterator:
    print(number)

```

```
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent?

#Answer
def powers_of_2(exponent):
  """
  Generates the powers of 2 up to a given exponent.

  Args:
    exponent: The maximum exponent.

  Yields:
    The powers of 2.
  """
  for i in range(exponent + 1):
     yield 2 ** i

# Example usage
for power in powers_of_2(5):
  print(power)
```

```
#7. Implement a generator function that reads a file line by line and yields each line as a string?
#Answer

def read_file_line_by_line(file_path):
  """
  Reads a file line by line and yields each line as a string.

  Args:
    file_path: The path to the file.

  Yields:
    Each line of the file as a string.
  """
  try:
    with open(file_path, 'r') as file:
      for line in file:
        yield line.strip()  # Strip leading/trailing whitespace
  except FileNotFoundError:
    print(f"Error: File '{file_path}' not found. Please make sure the file exists and the path is correct.")

# Example usage
file_path = 'my_file.txt'  # Replace with your file path

# Check if the file exists before attempting to read it
import os
if os.path.exists(file_path):
    for line in read_file_line_by_line(file_path):
        print(line)
else:
    print(f"Error: File '{file_path}' not found. Please create the file or provide a valid path.")
```

```
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit?
#Answer
def celsius_to_fahrenheit(celsius):
    """Converts Celsius to Fahrenheit."""
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 10, 20, 30, 40]

# Use map() to convert to Fahrenheit
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

# Print the results
print("Temperatures in Fahrenheit:", fahrenheit_temps)
```

```
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string?
#Answer
def remove_vowels(text):
    """Removes all vowels (a, e, i, o, u) from a string."""
    vowels = "aeiouAEIOU"
    filtered_letters = filter(lambda letter: letter not in vowels, text)
    return "".join(filtered_letters)
    # Example usage
text = "Hello, World!"
result = remove_vowels(text)
print(result)  # Output: Hll, Wrld!
```
